```python
continents=visa_dataset['continent'].unique()
for i in continents:
    cond1=visa_dataset['continent']==i
    cond2=visa_dataset['case_status']=='Certified'
    cond3=visa_dataset['case_status']=='Denied'
    count=len(visa_dataset[cond1&cond2])
    den_count=len(visa_dataset[cond1&cond3])
    #print("the number of certified people in {} is {}".format(i,count))
    #print("the number of denied people in {} is {}".format(i,den_count))
    print("the number of certified and denied people in {} is {} & {} respectively".format(i,count,den_count))
```

```python
certified,denied=[],[]
continents=visa_dataset['continent'].unique()
for i in continents:
    cond1=visa_dataset['continent']==i
    cond2=visa_dataset['case_status']=='Certified'
    cond3=visa_dataset['case_status']=='Denied'
    count=len(visa_dataset[cond1&cond2])
    den_count=len(visa_dataset[cond1&cond3])
    certified.append(count)
    denied.append(den_count)
```

```python
pd.DataFrame(zip(certified,denied),
            columns=("certified","denied"),
            index=visa_dataset["continent"].unique())
```
